```python


import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
# Create time data
time_p = np.linspace(0, 10, 1000) # 10 days, 1000 points

period_p = 2.5 # The star's period in days
amplitude = 0.8 # How much the brightness changes
brightness_p = 1.0 + amplitude * np.sin(2 * np.pi * time_p / period_p)

# Create the plot
plt.figure(figsize=(10, 5))
plt.plot(time_p, brightness_p, 'gold', linewidth=2)
plt.title('Pulsating Variable Star', fontsize=16, fontweight='bold')
plt.xlabel('Time (days)')
plt.ylabel('Apparent Brightness')
plt.grid(True, alpha=0.3)
plt.ylim(0, 2.2)
plt.show()

```
