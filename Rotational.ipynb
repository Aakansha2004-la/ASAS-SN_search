```python


import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')

time_r = np.linspace(0, 10, 1000)
period_r = 3.2 # The star's rotation period in days
# Simulate a rotating starspot
# The spot causes a dip in brightness once per rotation
brightness_r = 1.0 - 0.15 * np.sin(2 * np.pi * time_r / period_r)**8 # Using a high power to make the dip sharper

np.random.seed(42) # For reproducibility
brightness_r += np.random.normal(0, 0.01, size=len(time_r)) # Small random noise


plt.figure(figsize=(10, 5))
plt.plot(time_r, brightness_r, 'orchid', linewidth=2)
plt.title('Rotational Variable (Starspot)', fontsize=16, fontweight='bold')
plt.xlabel('Time (days)')
plt.ylabel('Apparent Brightness')
plt.grid(True, alpha=0.3)
plt.ylim(0.8, 1.05)
plt.show()

```
