```python


import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
# Create time data
time_e = np.linspace(0, 10, 1000)
period_e = 2.5 # Orbital period in days

# Calculate the orbital phase
phase = (time_e / period_e) % 1.0

# Start with a constant brightness
brightness_e = np.ones_like(time_e)

# Create a primary eclipse (deeper) when star 2 passes in front of star 1
primary_eclipse_mask = (phase > 0.45) & (phase < 0.55)
brightness_e[primary_eclipse_mask] = 0.4 # Deep dip

# Create a secondary eclipse (shallower) when star 1 passes in front of star 2
secondary_eclipse_mask = (phase > 0.95) | (phase < 0.05)
brightness_e[secondary_eclipse_mask] = 0.7 # Shallow dip


plt.figure(figsize=(10, 5))
plt.plot(time_e, brightness_e, 'cyan', linewidth=2)
plt.title('Eclipsing Binary Star System', fontsize=16, fontweight='bold')
plt.xlabel('Time (days)')
plt.ylabel('Apparent Brightness')
plt.grid(True, alpha=0.3)
plt.ylim(0, 1.2)
plt.show()

```
